```python
#IMPORTING THE NECESSARY LIBRARIES

import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
import numpy as np
```

```python
# GETTIN TO KNOW WITH THE DATA
df = pd.read_csv('all_data.csv')
print(df.head())
```

```python
# RENAMING COLUMS FOR BETTER ACCESSIBILITY

df.rename(columns = {'Country': 'country', 'Year': 'year', 'Life expectancy at birth (years)': 'life_exp_years', 'GDP': 'gdp'}, inplace = True)

print(df.head())

# CHECKING THE DATA TYPES FOR FURTHER USE

print(df.dtypes)

df['gdp_int'] = df.gdp.apply(lambda x: int(x))
df['gdp_billion_dollars'] = df.gdp_int.apply(lambda x: x/1000000000)

print(df['gdp_int'])
print(df['gdp_billion_dollars'])

del df['gdp_int']
del df['gdp']

# print(df.head())

# OUR DATA IS CLEAN AND READY TO USE
```

```python
# CHECKING THE DIFFERENCE IN THE GDP FOR THE YEARS IN THE 6 NATIONS

# print(df.country.value_counts())

years=[2000,2001,2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015]

chile = df[df.country == 'Chile']
# print(chile)
china = df[df['country'] == 'China']
usa = df[df.country == 'United States of America']
zim = df[df['country'] == 'Zimbabwe']
mex = df[df['country'] == 'Mexico']
ger = df[df['country'] == 'Germany']

plt.figure(figsize=(10,6))
sns.set()
sns.set_style('white')
sns.set_palette('Pastel2')
sns.set_context('talk')
ax = plt.subplot()
ax.set_title('The GDPs compared between the countries')
plt.plot(years, chile['gdp_billion_dollars'], label='Chile')
plt.plot(years,china['gdp_billion_dollars'],label='China')
plt.plot(years,usa['gdp_billion_dollars'],label='United States of America')
plt.plot(years,zim['gdp_billion_dollars'],label='Zimbabwe')
plt.plot(years,mex['gdp_billion_dollars'],label='Mexico')
plt.plot(years,ger['gdp_billion_dollars'],label='Germany')
plt.legend()
plt.xlabel('Year')
plt.ylabel('GDP (Billion Dollars)')
plt.savefig('gdp_between_countries.png')
plt.show()
```

```python
# GDP FOR DIFERENT NATIONS
sns.set()
sns.set_palette('Pastel1')
sns.set_style('white')
sns.set_context('talk')

plt.figure(figsize=(20, 25))

ax1 = plt.subplot(3,2,1)
ax1.set_title('Chiles GDP over the last 15 years')
ax1.set_yticks([])
ax1 = plt.plot(years,chile['gdp_billion_dollars'],label='Chile')
plt.xlabel('Year')
plt.legend(loc='upper left')

ax2 = plt.subplot(3,2,2)
ax2.set_title('Chinas GDP over the last 15 years')
ax2.set_yticks([])
ax2 = plt.plot(years,china['gdp_billion_dollars'],label='China')
plt.xlabel('Year')
plt.legend()
               
ax3= plt.subplot(3,2,3)
ax3.set_title('The USA\'s GDP over the last 15 years')
ax3.set_yticks([])               
ax3 = plt.plot(years,usa['gdp_billion_dollars'],label='United States of America')
plt.xlabel('Year')
plt.legend()
               
ax4 = plt.subplot(3,2,4)
ax4.set_title('Zimbabwes GDP over the last 15 years')
ax4.set_yticks([])
ax4 = plt.plot(years,zim['gdp_billion_dollars'],label='Zimbabwe')
plt.xlabel('Year')
plt.legend()
               
ax5 = plt.subplot(3,2,5)
ax5.set_title('Mexicos GDP over the last 15 years')
ax5.set_yticks([])
ax5 = plt.plot(years,mex['gdp_billion_dollars'],label='Mexico')
plt.xlabel('Year')
plt.legend()
               
ax6 = plt.subplot(3,2,6)
ax6.set_title('Germanys GDP over the last 15 years')
ax6.set_yticks([])
ax6 = plt.plot(years,ger['gdp_billion_dollars'],label='Germany')
plt.xlabel('Year')
plt.legend()
plt.savefig('gdp_countries.png')
plt.show()

```
